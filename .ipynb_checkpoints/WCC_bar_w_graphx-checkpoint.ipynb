```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([2, 4, 6])

# Modified cache size: YT: 2 MB, RN: 1 MB, LJ: 40 MB
# RN', 'LJ', 'YT'
cave = [1.843864, 1.481323, 0.099147]
graph_chi = [1568.583333, 67.49583333, 16.21656667]
grid_graph = [41.750706, 2.389181, 0.369182]
mosaic = [28.700520, 5.693491, 0.662670]
graphx = [629.4376149, 29.17456249, 5.906458738]

output_path = 'plot/wcc/wcc_bar_w_graphx'
```

```python
width = 0.2

# Graph Settings
fig = plt.figure(figsize=(60,30))
# plt.title(title,fontsize=52)


    
# plt.xlabel("x_label",fontsize=150)
plt.ylabel("Runtime (s)",fontsize=140)

plt.rc('xtick', labelsize = 120)    # fontsize of the tick labels
plt.rc('ytick', labelsize = 120)    # fontsize of the tick labels
plt.rc('legend', fontsize = 85)    # legend fontsize
plt.rc('lines', linewidth = 5)     # line width
plt.rc('lines', markersize = 60)   # marker size
plt.rc('lines', markeredgewidth = 4)   # marker edge width
plt.rc('lines', markerfacecolor = 'none') 

plt.rcParams["axes.linewidth"] = 5    #Border width

# plt.xticks(x,thread)
# plt.xticks(x, ['com-youtube', 'roadNet-PA', 'soc-LiveJournal1', 'BA_5000000', 'com-friendster']) 
plt.xticks(x, ['RN', 'LJ', 'YT']) 
# plt.xticks(x,num_fsize)
plt.yscale('log')
plt.tick_params(direction='out', length=40, width=3)    # Change tick size
plt.tick_params(direction='out', which='minor',length=30, width=3)

# Plot lines
plt.bar(x-0.4, graph_chi, width, edgecolor ='blue', color='None', linewidth = 5, hatch='X', label="GraphChi")
plt.bar(x-0.2, graphx, width, edgecolor ='darkorange', color='None', linewidth = 5, hatch='-', label="GraphX")
plt.bar(x, grid_graph, width, edgecolor ='red', color='None', linewidth = 5, hatch='/', label="GridGraph")
plt.bar(x+0.2, mosaic, width, edgecolor ='green', color='None', linewidth = 5, hatch='\\', label="Mosaic")
plt.bar(x+0.4, cave, width, edgecolor ='black', color='None', linewidth = 5, hatch='.', label="CAVE")

sz = 72
# for i in range(len(x)):
#     if i == 0:
#         plt.text(x[i]-0.1, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size=sz)
#         plt.text(x[i]+0.1, grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)
#     else:
#         plt.text(x[i]-0.1, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size=sz)
#         plt.text(x[i]+0.1, grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)

for i in range(len(x)):
    if i == 0:
        plt.text(x[i]-0.4, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size= sz)
        plt.text(x[i]-0.2, graphx[i], int(graphx[i]), ha="center", va="bottom", size=sz)
        plt.text(x[i], grid_graph[i], int(grid_graph[i]), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.2, mosaic[i], int(mosaic[i]), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.4, cave[i], round(cave[i],1), ha="center", va="bottom", size=sz)
    elif i == 1:
        plt.text(x[i]-0.4, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size= sz)
        plt.text(x[i]-0.2, graphx[i], int(graphx[i]), ha="center", va="bottom", size=sz)
        plt.text(x[i], grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.2, mosaic[i], round(mosaic[i],1), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.4, cave[i], round(cave[i],1), ha="center", va="bottom", size=sz)
    else:
        plt.text(x[i]-0.4, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size= sz)
        plt.text(x[i]-0.2, graphx[i], round(graphx[i],1), ha="center", va="bottom", size=sz)
        plt.text(x[i], grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.2, mosaic[i], round(mosaic[i],1), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.4, cave[i], round(cave[i],1), ha="center", va="bottom", size=sz)
        

# plt.ylim(top = 1400)
plt.legend()


plt.savefig(output_path +'.pdf', bbox_inches = "tight")

# If output front is small, rerun this cell
```
