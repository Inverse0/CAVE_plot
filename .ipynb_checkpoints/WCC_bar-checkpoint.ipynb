```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(6)
# 4th cache size
# 'FS', 'RN', 'YT', 'LJ', 'SD'
# graph_chi = [1406.63/37.661143, 403.432/2.316762, 8.36991/0.099490, 45.30303333/1.431606, 590.8996667/26.324257]
# grid_graph = [67.674578/37.661143, 20.384267/2.316762, 0.238151/0.099490, 2.356412/1.431606, 15.974468/26.324257]

# Modified cache size: FS, SD, TW: 1 GB, YT: 2 MB, RN: 1 MB, LJ: 40 MB
# 'FS', 'RN', 'YT', 'LJ', 'SD', 'TW'
graph_chi = [1406.63/37.661143, 1568.583333/1.843864, 16.21656667/0.099147, 67.49583333/1.481323, 590.8996667/26.324257, 974.813/34.746991]
grid_graph = [67.674578/37.661143, 41.750706/1.843864, 0.369182/0.099147, 2.389181/1.481323, 15.974468/26.324257, 27.500838/34.746991]
mosaic = [262.155842/37.661143, 28.700520/1.843864, 0.662670/0.099147, 5.693491/1.481323, 184.488631/26.324257, 85.336006/34.746991]

output_path = 'plot/wcc/wcc_bar'
```

```python
width = 0.2

# Graph Settings
fig = plt.figure(figsize=(60,30))
# plt.title(title,fontsize=52)

# ax = fig.add_axes((0, 0, 1, 1))

# for axis in ['top', 'bottom', 'left', 'right']:
#     ax.spines[axis].set_linewidth(5)  # change width
    
# plt.xlabel("x_label",fontsize=150)
plt.ylabel("Speedup",fontsize=140)

plt.rc('xtick', labelsize = 115)    # fontsize of the tick labels
plt.rc('ytick', labelsize = 115)    # fontsize of the tick labels
plt.rc('legend', fontsize = 105)    # legend fontsize
plt.rc('lines', linewidth = 5)     # line width
plt.rc('lines', markersize = 60)   # marker size
plt.rc('lines', markeredgewidth = 4)   # marker edge width
plt.rc('lines', markerfacecolor = 'none') 

plt.rcParams["axes.linewidth"] = 5    #Border width

# plt.xticks(x,thread)
# plt.xticks(x, ['com-youtube', 'roadNet-PA', 'soc-LiveJournal1', 'BA_5000000', 'com-friendster']) 
plt.xticks(x, ['FS', 'RN', 'YT', 'LJ', 'SD', 'TW']) 
# plt.xticks(x,num_fsize)
plt.yscale('log')
plt.tick_params(direction='out', length=40, width=3)    # Change tick size
plt.tick_params(direction='out', which='minor',length=30, width=3)

# Plot lines
plt.bar(x-0.2, graph_chi, width, edgecolor ='blue', color='None', linewidth = 5, hatch='X', label="GraphChi")
plt.bar(x, grid_graph, width, edgecolor ='red', color='None', linewidth = 5, hatch='/', label="GridGraph")
plt.bar(x+0.2, mosaic, width, edgecolor ='green', color='None', linewidth = 5, hatch='\\', label="Mosaic")

sz = 74
# for i in range(len(x)):
#     if i == 0:
#         plt.text(x[i]-0.1, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size=sz)
#         plt.text(x[i]+0.1, grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)
#     else:
#         plt.text(x[i]-0.1, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size=sz)
#         plt.text(x[i]+0.1, grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)

for i in range(len(x)):
    if i == 1:
        plt.text(x[i]-0.2, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size= sz)
        plt.text(x[i], grid_graph[i], int(grid_graph[i]), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.2, mosaic[i], int(mosaic[i]), ha="center", va="bottom", size=sz)
    else:
        plt.text(x[i]-0.2, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size= sz)
        plt.text(x[i], grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.2, mosaic[i], round(mosaic[i],1), ha="center", va="bottom", size=sz)

plt.ylim(top = 1400)
plt.legend()


plt.savefig(output_path +'.pdf', bbox_inches = "tight")

# If output front is small, rerun this cell
```
