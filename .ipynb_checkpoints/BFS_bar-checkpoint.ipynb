```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(6) 
# 4th cache size
# 'FS', 'RN', 'YT', 'LJ', 'SD'
# graph_chi = [265.7646667/36.195423, 548.1993333/2.382215, 3.664293333/0.096749, 45.23756667/0.867602, 1018.319/26.080364]
# grid_graph = [123.821243/36.195423, 23.195117/2.382215, 0.424254/0.096749, 3.891406/0.867602, 27.457819/26.080364]


# Modified cache size: FS, SD, TW: 1 GB, YT: 2 MB, RN: 1 MB, LJ: 40 MB
# Order: 'FS', 'RN', 'YT', 'LJ', 'SD', 'TW'
graph_chi = [265.7646667/36.195423, 1990.973333/2.022302, 7.128166667/0.100472, 66.4868/0.863348, 1018.319/26.080364, 1080.533333/24.744072]
grid_graph = [123.821243/36.195423, 46.158397/2.022302, 0.749779/0.100472, 3.814387/0.863348, 27.457819/26.080364, 50.576340/24.744072]
mosaic = [187.879225/36.195423, 0.108813/2.022302, 0.183293/0.100472, 4.471696/0.863348, 113.357057/26.080364, 70.634405/24.744072]
```

```python
width = 0.2

# Graph Settings
fig = plt.figure(figsize=(60,36))
# plt.title(title,fontsize=52)

# ax = fig.add_axes((0, 0, 1, 1))

# for axis in ['top', 'bottom', 'left', 'right']:
#     ax.spines[axis].set_linewidth(5)  # change width
    
# plt.xlabel("x_label",fontsize=150)
plt.ylabel("Speedup",fontsize=140)

plt.rc('xtick', labelsize = 115)    # fontsize of the tick labels
plt.rc('ytick', labelsize = 115)    # fontsize of the tick labels
plt.rc('legend', fontsize = 105)    # legend fontsize
plt.rc('lines', linewidth = 5)     # line width
plt.rc('lines', markersize = 60)   # marker size
plt.rc('lines', markeredgewidth = 4)   # marker edge width
plt.rc('lines', markerfacecolor = 'none') 
# plt.rcParams['hatch.linewidth'] = 1

plt.rcParams["axes.linewidth"] = 5    #Border width

# plt.xticks(x,thread)
# plt.xticks(x, ['com-youtube', 'roadNet-PA', 'soc-LiveJournal1', 'BA_5000000', 'com-friendster']) 
plt.xticks(x, ['FS', 'RN', 'YT', 'LJ', 'SD', 'TW']) 
# plt.xticks(x,num_fsize)
plt.yscale('log')
plt.tick_params(direction='out', length=40, width=3)    # Change tick size
plt.tick_params(direction='out', which='minor',length=30, width=3)

# Plot lines
plt.bar(x-0.2, graph_chi, width, edgecolor ='blue', color='None', linewidth = 5, hatch='X', label="GraphChi")
plt.bar(x, grid_graph, width, edgecolor ='red', color='None', linewidth = 5, hatch='/', label="GridGraph")
plt.bar(x+0.2, mosaic, width, edgecolor ='green', color='None', linewidth = 5, hatch='\\', label="Mosaic")
# plt.bar(x+0.2, y3, width, edgecolor ='black', color='None', linewidth = 5, label="y3_name")

sz = 74
gap = 1
# for i in range(len(x)):
#     if i == 0:
#         plt.text(x[i]-0.1, graph_chi[i], round(graph_chi[i],1), ha="center", va="bottom", size= sz)
#         plt.text(x[i]+0.1, grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)
#     else:
#         plt.text(x[i]-0.1, graph_chi[i], int(graph_chi[i]), ha="center", va="bottom", size=sz)
#         plt.text(x[i]+0.1, grid_graph[i], round(grid_graph[i],1), ha="center", va="bottom", size=sz)
for i in range(len(x)):
    if i == 1:
        plt.text(x[i]-0.2, graph_chi[i]*gap, int(graph_chi[i]), ha="center", va="bottom", size= sz)
        plt.text(x[i], grid_graph[i]*gap, int(grid_graph[i]), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.2, mosaic[i]*gap, round(mosaic[i],1), ha="center", va="bottom", size=sz)
    else:
        plt.text(x[i]-0.2, graph_chi[i]*gap, int(graph_chi[i]), ha="center", va="bottom", size=sz)
        plt.text(x[i], grid_graph[i]*gap, round(grid_graph[i],1), ha="center", va="bottom", size=sz)
        plt.text(x[i]+0.2, mosaic[i]*gap, round(mosaic[i],1), ha="center", va="bottom", size=sz)
        
plt.ylim(top = 1800)

plt.legend()


plt.savefig('plot/bfs/bfs_bar' +'.pdf', bbox_inches = "tight")

# If output front is small, rerun this cell
```
